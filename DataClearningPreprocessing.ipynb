we are imporing sammple text form a text file in my local computer file name is sampletext.txt

```python
# Open the file and read the content
with open('sampletext.txt', 'r', encoding='utf-8') as file:
    sent1 = file.read()

# Now, sent1 contains the text from the file
print(sent1)
```

Remove Punctuation 

```python
import string
translator = str.maketrans('', '', string.punctuation)
sent_pun = sent1.translate(translator)
print(sent_pun)
```

Remove Unicode

```python
sent_pun_uni = sent_pun.encode('ascii', 'ignore').decode("utf-8")
print(sent_pun_uni)
# Output: Feeling loved even when Im sicklikeforlike chocolate bf iloveyou aftereight couplegoals
```

Remove URL

```python
import re
subject = "our web site offers it courses online https://www.industryonboarding.com"
result = re.sub(r"http\S+", "", subject)
print(result)
# Output: Omg, check out these fabulous shoes  yes they can be yours
```

```python
from nltk.corpus import stopwords

# List of English stopwords
stop_en = stopwords.words("english")
# Example sentence broken down into words (tokens)
sentence = ["we", "went", "to", "the", "park", "to", "play", "and", "had", "a", "lot", "of", "fun"]
# Removing stopwords from the sentence
cleaned_sentence = [word for word in sentence if word not in stop_en]
print(cleaned_sentence)
# Output: ['went', 'park', 'play', 'lot', 'fun']
```
